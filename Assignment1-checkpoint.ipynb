```python
#Removing redundant white spaces from whole data frame string vlaues

#Changing all the string data to upper case

#changing the ranking columns from float to int data type as the rankis are integer values

for i in (starwars.columns):
    if starwars[i].dtype==object:
        starwars[i]=starwars[i].str.strip()
        starwars[i]=starwars[i].str.upper()
    if starwars[i].dtype==float:
        starwars[i]=starwars[i].astype('Int64')
```

```python
#Task 2: Data Exploration


#2.1:  Exploraiton of the survey question: Rank of the star wars films


#Assigning unique value counts of columns 10 to 15 data to an object and printing the object

Ratings=pd.DataFrame(starwars.iloc[:,9:15].apply(pd.Series.value_counts))
Ratings.insert(0,'rating',[1,2,3,4,5,6],True)
Ratings.set_index('rating',inplace=True)
Ratings
```

```python
#To change the view of plot in between inline and outline

#%matplotlib qt
%matplotlib inline
```

```python
#Visualisation for ratings of the movies across movie names

import numpy as np
import matplotlib.pyplot as plt
 
bar_width=0.1

# Set position of bar on X axis
r1 = np.arange(len(Ratings))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]
r5 = [x + bar_width for x in r4]
r6 = [x + bar_width for x in r5]

plt.figure(figsize=(10,5))


# Make the plot
plt.bar(r1, Ratings.iloc[:,0], color='b', width=bar_width,alpha=0.25)
plt.bar(r2, Ratings.iloc[:,1], color='g', width=bar_width,alpha=0.25)
plt.bar(r3, Ratings.iloc[:,2], color='r', width=bar_width,alpha=0.25)
plt.bar(r4, Ratings.iloc[:,3], color='c', width=bar_width,alpha=0.25)
plt.bar(r5, Ratings.iloc[:,4], color='m', width=bar_width,alpha=0.25)
plt.bar(r6, Ratings.iloc[:,5], color='y', width=bar_width,alpha=0.25)
 
# Add xticks on the middle of the group bars
plt.title('Starwars Franchise Movies Rankings',fontsize='20')
plt.xlabel('Rankings',fontsize='16')
plt.ylabel('Frequency of Rankings', fontweight='bold',fontsize='16')
plt.xticks([r1 + bar_width for r1 in range(len(Ratings['rank_of_Episode_I_The_Phantom_Menace']))],Ratings.index)
 
# Create legend & Show graphic
plt.legend(['Movie1', 'Movie2', 'Movie3','Movie4','Movie5','Movie6'], loc='upper center',fontsize='small',ncol=3)
plt.grid()
plt.show()
```

```python
#Visualisation for View of Starwar characters by Gender

#Generating cross tabulation for Starwars charcaters and Gender usinf for loop and generating the plot

for val in starwars.iloc[:,15:29]:
    data=pd.crosstab(index=starwars[val],
            columns=starwars['Gender']).reset_index()
    bar_width=0.1
    # Set position of bar on X axis
    r1 = np.arange(len(data))
    r2 = [x + bar_width for x in r1]
    r3 = [x + bar_width for x in r2]
    r4 = [x + bar_width for x in r3]
    r5 = [x + bar_width for x in r4]
    r6 = [x + bar_width for x in r5]
   
    # Make the plot
    plt.figure(figsize=(10,5))
    
    plt.bar(r1, data.iloc[:,1], color='r', width=bar_width,alpha=0.25)
    plt.bar(r2, data.iloc[:,2], color='b', width=bar_width,alpha=0.25)
    
    # Add xticks on the middle of the group bars
    plt.title(data.columns[0],fontsize='16')
    plt.xlabel('Gender',fontsize='12')
    plt.ylabel('Frequency of views', fontweight='bold',fontsize='12')
    plt.xticks([r1 + bar_width for r1 in range(len(data))],['Neutral','Some_Fav','Some_Un_Fav','Un_familiar','Very_Fav','Very_Un_Fav'])
    # Create legend & Show graphic
    plt.legend(data.columns[1:3], loc='upper center',fontsize='small',ncol=3)
    plt.grid()
    plt.show()
```
